```python
#Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Stages 10 and above not included

input2_csv = "/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Area_2h_firsttimepoints_for xcal.csv"

df2 = pd.read_csv(input2_csv)

# Plotting setup

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))


sns.swarmplot(x='Time3_no10s', y='Percent_areaincreaseBAck_no10s', data=df2, color='grey', hue='EC_67_15updated_n10', marker='o', s=4, edgecolor='gray', linewidth=0.15)
ax = sns.boxplot(x='Time3_no10s', y='Percent_areaincreaseBAck_no10s', data=df2, palette='viridis', linewidth=1)



plt.ylabel('Percent of egg chamber area', fontsize=10)
plt.xlabel('Egg chambers', fontsize=10)
plt.yticks(size=10)
plt.xticks(size=10)
plt.legend().set_visible(False)

#plt.legend(fontsize=5)
plt.legend().set_visible(False)

#plt.xlim(-0.5, 6.5)


# Add a frame around the plot
frame = plt.gca().add_patch(plt.Rectangle((plt.xlim()[0], plt.ylim()[0]), plt.xlim()[1] - plt.xlim()[0], plt.ylim()[1] - plt.ylim()[0], fill=None, edgecolor='black', linewidth=1))

# Set line width for each side of the frame
frame.set_linewidth(1)
frame.set_edgecolor('grey')


plt.savefig('/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/boxPercent_growthchange_fineX.pdf', dpi=300, bbox_inches='tight') 

plt.show() 
```
