```python

# Drop rows with NaN values
d_clean = d.dropna(subset=['InVial', 'On Bench', 'On Mcroscope', 'On Laser'])

# Define data
InVial = d_clean['InVial']
OnBench = d_clean['On Bench']
OnMcroscope = d_clean['On Mcroscope']
OnLaser = d_clean['On Laser']

# Perform t-test between InVial and OnBench
t_statistic, p_value = stats.ttest_ind(InVial, OnBench, equal_var=False)  # Assuming unequal variances

print("T-Statistic:", t_statistic)
print("P-value:", p_value)

# Perform t-test between InVial and OnMcroscope
t_statistic, p_value = stats.ttest_ind(InVial, OnMcroscope, equal_var=False)  # Assuming unequal variances

print("T-Statistic:", t_statistic)
print("P-value:", p_value)

# Perform t-test between InVial and OnLaser
t_statistic, p_value = stats.ttest_ind(InVial, OnLaser, equal_var=False)  # Assuming unequal variances

print("T-Statistic:", t_statistic)
print("P-value:", p_value)
```
