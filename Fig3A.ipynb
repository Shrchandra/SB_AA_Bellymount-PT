```python
## Import libraries###

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score 


# Interapolating data
input_csv = "/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Area_comparision.csv"
output_csv = "/Users/shruthibalachandra/Area_intrapolated_updated.csv"

df = pd.read_csv(input_csv)

# Interpolate missing values using polynomial interpolation

df['Int_King 1970'] = df['King 1970'].interpolate(methods='polynomial', order=5)
df['Int_Spardling'] = df['Spardling'].interpolate(methods='polynomial', order=5)
df['Int_67_15_updated'] = df['67_15_updated'].interpolate(methods='polynomial', order=5)

# Save interpolated data to a new CSV file
df.to_csv(output_csv, index=False)


# Plotting
plt.figure(figsize = (4,3))
plt.style.use('seaborn')
sns.set_style("whitegrid")

# Plot original data and interpolated data for King 1970
sns.lineplot(x='Hours', y='King 1970', color='g', data=df, marker='o', label='King 1970 (Original)')
sns.lineplot(x='Hours', y='Int_King 1970', data=df, color='g', dashes=True, marker='d', alpha=0.5, label='King 1970 (Interpolated)')

# Plot original data and interpolated data for Spardling
sns.lineplot(x='Hours', y='Spardling', color='#0068a6', data=df, marker='o', label='Spardling (Original)')
sns.lineplot(x='Hours', y='Int_Spardling', data=df, color='#0068a6', dashes=True, marker='d', alpha=0.5, label='Spardling (Interpolated)')

# Plot original data and interpolated data for 67_15_updated
sns.lineplot(x='Hours', y='67_15_updated', color='k', data=df, marker='o', label='67_15_updated (Original)')
sns.lineplot(x='Hours', y='Int_67_15_updated', data=df, color='k', dashes=True, marker='d', alpha=0.5, label='67_15_updated (Interpolated)')

plt.xlabel('Hours', fontsize=12)
plt.ylabel('Area', fontsize=12)
plt.title('Interpolated Data Comparison', fontsize=14)
plt.legend()
plt.grid(True)

#Saving interapolating data_all genotype

input4_csv = "/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Area_intrapolated_updated.csv"

plt.show()
```

```python
# King data

input2_csv = "/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Area_intrapolated_edited_King_forcurve2.csv"

df2 = pd.read_csv(input2_csv)


# Define data

x2 = df2['Hours2']
y2 = df2['Int_DavidMarley']


# Fit a curve using polynomial regression (linear regression on log-transformed data)

p2 = np.polyfit(x2, np.log(y2), 1)
f2 = lambda x2: np.exp(p2[1]) * np.exp(p2[0]*x2)


# calculate R-squared value

R2_square = r2_score(y2, f2(x2)) 

print('Coefficient of Determination', R2_square)
```
