```python
### Import libraries ###

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats

#### Import data ###

d=pd.read_csv('/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Imaging_conditions_status_v34.csv')

d.head()

```

```python

# Defining fig size
plt.figure(figsize = (2,3))
sns.set_style("whitegrid")

# Define custom color palettes
custom_palette1 = ["#006ca5"]
custom_palette2 = ["#20B2AA"]

# Convert 'ConditionsDayA' and 'ConditionsDayD' columns to strings
d['ConditionsDayA'] = d['ConditionsDayA']
d['ConditionsDayD'] = d['ConditionsDayD']

# Define the data
conditions = ['Foodvial', 'Bench', 'Microscope', 'Imaged']
average_a = [100, 72.38095238, 65.95744681, 58.62068966]
average_d = [0, 27.61904762, 34.04255319, 41.37931034]


# Plot bars for AverageA
plt.bar(conditions, average_a, color=custom_palette1[0], label='Alive')

# Plot bars for AverageD on top of AverageA
plt.bar(conditions, average_d, bottom=average_a, color=custom_palette2[0], label='Dead')


# Additional plot customizations
plt.xlabel('Imaging conditions', fontsize=10)
plt.ylabel('Percent of flies', fontsize=10)
plt.xticks(size=10, rotation=35, rotation_mode='anchor', ha='right')
plt.yticks(size=10)
plt.legend(['Alive', 'Dead'])

# Save the plot as a PNG file
plt.savefig('/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/BM_Figure_Plots/Status_BM_Alive_Dead_stacked.pdf', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()

```
