```python
############# Fig5E ##################

#Importing libraries

import numpy as np
from sklearn.metrics import r2_score
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the live imaging data
input_csv_live = "/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/All_FCSC_normalized_reordered.csv"
df_live = pd.read_csv(input_csv_live)

# Load the fixed imaging data
input_csv_fixed = "/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/His_Fixed_timecropped75h.csv"
df_fixed = pd.read_csv(input_csv_fixed)


# Extract the relevant columns from the DataFrame
y6 = df_fixed['NCI_time']
x6 = df_fixed['timeNCI']

# Ensure there are no NaN or infinite values in x and y
valid_indices = np.isfinite(x6) & np.isfinite(y6)
x6 = x6[valid_indices]
y6 = y6[valid_indices]

# Perform the log transformation on y and fit a polynomial (linear) to the data
p6 = np.polyfit(x6, np.log(y6), 1)

# Define the exponential function to model the data
f6 = lambda x6: np.exp(p6[1]) * np.exp(p6[0]*x6)

# Calculate the R-squared value
y6_log_pred = f6(x6)
R6_square = r2_score(np.log(y6), np.log(y6_log_pred))

# Extract the slope and intercept
slope, intercept = p6

print('Coefficient of Determination (R):', R6_square)
print('Intercept:', intercept)
print('Slope:', slope)
```
