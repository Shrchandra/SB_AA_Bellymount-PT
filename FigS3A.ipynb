```python

#Libraries
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker

# Read the CSV file into a DataFrame
d=pd.read_csv('/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Stage_comparison.csv')


# Set up plot aesthetics
plt.style.use('seaborn')
sns.set_style("whitegrid")

#defining fig size
plt.figure(figsize=(6,4))


# Plot stripplot and boxplot
sns.stripplot(x='Stages', y='67_15_updated', data=d, color='grey', marker='o', s=2, edgecolor='black', linewidth=0.15)
ax = sns.boxplot(x='Stages', y='67_15_updated', data=d, palette='rainbow', linewidth=0.75)

# Set labels and ticks
plt.xlabel('Developmental stages', fontsize=10)
plt.ylabel(' Egg chamber area (μm^2)', fontsize=10)
plt.xticks(size=10, rotation=35, rotation_mode = 'anchor', ha='right')
plt.yticks(size=10)

# Set y-axis to logarithmic scale
plt.yscale("log")


# Add a frame around the plot
frame = plt.gca().add_patch(plt.Rectangle((plt.xlim()[0], plt.ylim()[0]), plt.xlim()[1] - plt.xlim()[0], plt.ylim()[1] - plt.ylim()[0], fill=None, edgecolor='black', linewidth=1))
frame.set_linewidth(1)
frame.set_edgecolor('grey')

# Customize y-axis ticks

yticks = [1000, 2500, 5000, 7500, 10000, 25000, 50000, 75000]  # Adjust this according to your data range
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
ax.set_yticks(yticks)


# Save the plot as a PDF file
#plt.savefig('/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Boxplot_age_log_updated2.pdf', dpi=300, bbox_inches='tight') 


# Display the plot    
plt.show() 
```
