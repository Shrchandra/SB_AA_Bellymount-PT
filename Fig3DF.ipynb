```python
## Import libraries

import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import r2_score

# Set seaborn figure style
sns.set(rc={'figure.figsize': (4, 3), 'font.family': 'sans-serif', 'font.sans-serif': "Arial"})

# Load data
input1_csv = "/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/67_15_updatecurve2.csv"
input2_csv = "/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Area_2h_firsttimepoints_for xcal.csv"
input4_csv = "/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Updated_6715_finetimeing_stage10.csv"


df1 = pd.read_csv(input1_csv)
df2 = pd.read_csv(input2_csv)
df4 = pd.read_csv(input4_csv)

# Extract data
x = df1['Hours6']
y = df1['Int_67_15_updated']

# Fit a curve using linear regression on log-transformed data
p = np.polyfit(x, np.log(y), 1)
f = lambda x: np.exp(p[1]) * np.exp(p[0]*x)

# Calculate R-squared value
R_square = r2_score(y, f(x))
print('Coefficient of Determination:', R_square)
print('Slope:', p[0])
print('Intercept (exp(p[1])):', np.exp(p[1]))
print('Slope (exp(p[0])):', np.exp(p[0]))


# Plotting

plt.figure(figsize=(4, 3))
plt.style.use('seaborn')
sns.set_style("whitegrid")

x = df1['Hours6']

plt.plot(x, f(x), color='k', label='Curve Fit')
plt.plot(x, y, color='k', linestyle='dashed', label='Actual Data')


plt.yscale("log")
plt.xlabel('Developmental time(h)', fontsize=10)
plt.ylabel('Egg chamber area (Î¼m^2)', fontsize=10)

plt.xticks(size=10)
plt.yticks(size=10)
plt.xlim(0, 95)

# Save and display the plot
#plt.savefig('/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/67_15growthrate_std.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
