```python

# Define the data
conditions = ['Foodvial', 'Bench', 'Microscope', 'Imaged']

# Percentages for each interval and condition
alive_10min = [100, 80, 0, 0]
dead_10min = [0, 20, 100, 100]
alive_1h = [100, 75, 83.33, 80]
dead_1h = [0, 25, 16.67, 20]
alive_2h = [100, 72.38, 65.96, 58.62]
dead_2h = [0, 27.62, 34.04, 41.38]

# Define color palettes for each interval
color_10min = "#006ca5"  # Light salmon for 10min
color_1h = "#20B2AA"     # Light sea green for 1h
color_2h = "#cc5500"      # Medium purple for 2h

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))

# Bar width and positions
bar_width = 0.25
x = range(len(conditions))

# Plotting for each interval with differentiated colors
# 10min interval
ax.bar(x, alive_10min, width=bar_width, color=color_10min, label='Alive (10min)')
ax.bar(x, dead_10min, width=bar_width, bottom=alive_10min, color=color_10min, alpha=0.5, label='Dead (10min)')

# 1h interval
ax.bar([p + bar_width for p in x], alive_1h, width=bar_width, color=color_1h, label='Alive (1h)')
ax.bar([p + bar_width for p in x], dead_1h, width=bar_width, bottom=alive_1h, color=color_1h, alpha=0.5, label='Dead (1h)')

# 2h interval
ax.bar([p + bar_width*2 for p in x], alive_2h, width=bar_width, color=color_2h, label='Alive (2h)')
ax.bar([p + bar_width*2 for p in x], dead_2h, width=bar_width, bottom=alive_2h, color=color_2h, alpha=0.5, label='Dead (2h)')

# Customize the plot
ax.set_xlabel('Imaging Conditions', fontsize=10)
ax.set_ylabel('Percent of Flies', fontsize=10)
ax.set_title('Percent of Flies Alive and Dead Across Different Imaging Intervals', fontsize=12)
ax.set_xticks([p + bar_width for p in x])
ax.set_xticklabels(conditions, ha='center')

# Legend

plt.legend([])

# Save the plot as a PNG file
plt.savefig('/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/BM_Figure_Plots/Status_BM_Alive_Dead_stacked_intervals.pdf', dpi=300, bbox_inches='tight')

plt.show()
```
