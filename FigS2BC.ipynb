```python
#Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot 
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import pearsonr
# Load data
input10_csv = "/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Degen_0h_6h.csv"
df10 = pd.read_csv(input10_csv)

#Plotting
plt.style.use('seaborn')
sns.set_style("whitegrid")
plt.rcParams.update({'figure.figsize': (4, 3), 'font.size': 30, 'font.family': 'sans-serif', 'font.sans-serif': "Arial"})

# Extract x and y data
#x_a = df10['0h_alive']
#y_a = df10['6h_alive']

x_d = df10['0h_dead']
y_d = df10['6h_dead']

# Calculate Pearson correlation
corr_coefficient, p_value = pearsonr(x_d, y_d)
print(f"Pearson correlation coefficient: {corr_coefficient}")
print(f"P-value: {p_value}")
# Scatter plot with linear regression line
sns.regplot(x=x_d, y=y_d, color='#20B2AA', data=df10, ci=None, scatter_kws={'s': 60})

# Labels and legend
plt.xlabel('0h of imaging', fontsize=10)
plt.ylabel('6h of imaging', fontsize=10)
plt.legend()
plt.xlim(-10, 50)
plt.ylim(-10, 65)

# Save and display plot
plt.savefig('/Users/shruthibalachandra/Documents/Shruthi_for_mac/Amodeo_Lab/Python/Degen_correlation.pdf', dpi=300, bbox_inches='tight') 
plt.show()
```
